Covid 19 cases among passengers of international flight arriving at Japanese airports   

```python
import pandas as pd
```

```python
df = pd.read_csv('cases_airport_jp.csv', index_col='id')
df
```

```python
# df.info()
```

Change the year of rows where the date is November or December to 2020. 

```python
# find rows where date is November or December
filt = (df['date'].str.startswith('12/')) | (df['date'].str.startswith('11/'))

# df.loc[filt]

# Change the year of these columns to 2020
df.loc[filt, ['year']] = 2020
```

```python
# check if the year has been changed
df.loc[filt]
```

```python
# sort df by case_id
df = df.sort_values(by='case_id')
```

After checking the original report, case_id 307 is actually 3071. 

Correct the case_id from 307 to 3071

```python
filt = df['case_id'] == 307
df.loc[filt, ['case_id']] = 3071
```

```python
df[df['case_id'] == 3071]
```

```python
df.sort_values(by='case_id')
```

Change data in column 'date' to datetime

```python
df.info()
```

```python
# add year to the date
def combine_date_year(row):
    return row['date'] + '/' + str(row['year'])
```

```python
df['date'] = df[['date', 'year']].apply(combine_date_year, axis=1)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

```python
df = df.sort_values(by='date')
```

Save cleaned data as csv

```python
df.to_csv('cases_airport_jp_cleaned.csv')
```
